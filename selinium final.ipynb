```python
#salary

avg_salary=driver.find_elements_by_xpath("//div[@class='col-12 col-lg-4 px-lg-0 d-flex align-items-baseline']")
avg_salary


Avg_salary=[]
for i in avg_salary:
    Avg_salary.append(i.text.replace("\n",""))
Avg_salary
```

```python
#rating

job_rating=driver.find_elements_by_xpath("//span[@class='m-0 css-kyx745']")  
job_rating

Job_rating=[]
for i in job_rating:
    Job_rating.append(i.text)
Job_rating
```

```python
min_salary=[]
max_salary=[]



for i in page_url:
    driver.get(i)
    time.sleep(3)
    
    try:
        min=driver.find_element_by_xpath("//div[@class='d-flex flex-column css-12efcmn css-fajoci eb8fmw6']/strong")
        min_salary.append(min.text)
    except:
        min_salary.append('-')
        
        
    try:
        max=driver.find_element_by_xpath("//div[@class='d-flex flex-column css-12efcmn align-items-end css-p6bucf eb8fmw6']/strong")
        max_salary.append(max.text)
    except:
        max_salary.append('-')
        
        
        
print(min_salary)
print(max_salary)
```
