# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
page=requests.get("https://www.imdb.com/chart/top")
page

page.content

soup=BeautifulSoup(page.content)
soup

title_name=soup.find_all("td",class_="titleColumn")
title_name

movie_name=[]
for i in title_name:
    for j in i.find_all('a'):
        movie_name.append(j.text.replace("\n",""))
        
movie_name

rating=soup.find_all("td",class_="ratingColumn imdbRating")
rating

movie_rating=[]
for i in rating:
    movie_rating.append(i.text.replace("\n",""))
        
movie_rating

year=soup.find_all("span",class_="secondaryInfo")
year

movie_year=[]
for i in year:
    movie_year.append(i.text.replace("\n",""))
        
movie_year

df=pd.DataFrame({})
df['movie_name']=movie_name[:100]
df['movie_rating']=movie_rating[:100]
df['movie_year']=movie_year[:100]
df



```

# 8.) Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
page=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YScqrUvivIU")
page

soup=BeautifulSoup(page.content)
soup

period_time=soup.find_all("div",class_="col-sm-2 forecast-label")
period_time

Period_name=[]
for i in period_time:
    i=i.text.replace("\n","")
    i=i.strip(" ")
    Period_name.append(i)
Period_name=Period_name[1::2]        
Period_name


short_des=soup.find_all("div",class_="col-sm-10 forecast-text")
short_des

Description=[]
for i in short_des:
    i=i.text.replace("\n","")
    i=i.strip(" ")
    i=i.split('.')[0]
    Description.append(i)

Description=Description[1::2]       
Description


temp=soup.find_all("div",class_="col-sm-10 forecast-text")
temp

Temp=[]
for i in temp:
    i=i.text.replace("\n","")
    i=i.strip(" ")
    i=i.split(', with a')[1].split('.')[0][-2:]
    Temp.append(i)
        
Temp=Temp[1::2]
Temp


full_des=soup.find_all("div",class_="col-sm-10 forecast-text")
full_des

Full_des=[]
for i in full_des:
    i=i.text.replace("\n","")
    i=i.strip(" ")
    Full_des.append(i)

Full_des=Full_des[1::2]       
Full_des


df=pd.DataFrame({})
df['Period_name']=Period_name
df['Description']=Description
df['Temp']=Temp
df['Full_des']=Full_des
df

```
