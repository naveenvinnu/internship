# Modeling Using Techniques:

```python
x=df.copy()
x.drop('IndianOnlineRetailerToRecommend',axis=1,inplace=True)
y=df['IndianOnlineRetailerToRecommend']
```

```python
cat=[i for i in x.columns if x[i].dtypes=='O']
```

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
encode=OrdinalEncoder()
labe=LabelEncoder()
```

```python
#using ordinal encoder for independent features
for i in cat:
    x[i]=encode.fit_transform(x[i].values.reshape(-1,1))

#Using label encoder for Label Column
y=labe.fit_transform(y)
```

# Scaling Method:

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
xt=scaler.fit_transform(x)
x=pd.DataFrame(xt,columns=x.columns)
```

```python
from sklearn.ensemble import RandomForestClassifier
m=RandomForestClassifier()
m.fit(x,y)
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(m.feature_importances_, index=x.columns)
plt.figure(figsize=(10,8))
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

Observations:

In the above chart we can see that above features are of most importance in determining which platform will a customer recommend to his friend.

Using chi2 test

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
selection = SelectKBest(score_func=chi2)
fit = selection.fit(x,y)
```

```python
dtscores = pd.DataFrame(fit.scores_)
dtcolumns = pd.DataFrame(x.columns)
featureScores = pd.concat([dtcolumns,dtscores],axis=1)
featureScores.columns = ['Features','Score']
```

```python
print(featureScores.nlargest(15,'Score'))  #print10 best features
feat=list(featureScores.nlargest(12,'Score')['Features'])
```

```python
from sklearn.model_selection import train_test_split,cross_val_score

from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
model=RandomForestClassifier()
model.fit(xtrain,ytrain)
p=model.predict(xtest)
s=cross_val_score(model,x,y,cv=10)
```

```python
print('Accuracy',np.round(accuracy_score(p,ytest),4))
print('----------------------------------------------------------')
print('Mean of Cross Validation Score',np.round(s.mean(),4))
print('----------------------------------------------------------')
print('Confusion Matrix')
print(confusion_matrix(p,ytest))
print('----------------------------------------------------------')
print('Classification Report')
print(classification_report(p,ytest))
```

```python
model=XGBClassifier(verbosity=0)
model.fit(xtrain,ytrain)
p=model.predict(xtest)
s=cross_val_score(model,x,y,cv=10)
```

```python
print('Accuracy',np.round(accuracy_score(p,ytest),4))
print('----------------------------------------------------------')
print('Mean of Cross Validation Score',np.round(s.mean(),4))
print('----------------------------------------------------------')
print('Confusion Matrix')
print(confusion_matrix(p,ytest))
print('----------------------------------------------------------')
print('Classification Report')
print(classification_report(p,ytest))
```
