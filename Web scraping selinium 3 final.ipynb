# 4.Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver=webdriver.Chrome('chromedriver.exe')
url='https://www.flipkart.com'
driver.get(url)
```

```python
search_product=driver.find_element_by_xpath("//input[@title='Search for products, brands and more']")
search_product
```

```python
search_product.send_keys('samsung a12')
```

```python
search_button=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_button.click()
```

```python
Mobile_Name=[]
Brand_Name=[]
Url=[]
mn=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in mn:
    Mobile_Name.append(i.text)
    
bn=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in bn:
    Brand_Name.append(i.text[0:5])

url=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in url:
    Url.append(i.get_attribute('href'))
```

```python
Model_No=[]
Color=[]
Price=[]

for i in Url:
    driver.get(i)
    m=driver.find_element_by_xpath("//div[@class='_1AtVbE col-12-12']//table[1]//tr[2]//td[2]//li[1]")
    Model_No.append(m.text)
    
    clr=driver.find_element_by_xpath("//div[@class='_1AtVbE col-12-12']//table[1]//tr[4]//td[2]//li[1]")
    Color.append(clr.text)
    
    prc=driver.find_element_by_xpath("//div[@class='_25b18c']//div[1]")
    Price.append(prc.text)
```

```python
dis=[]
res=[]
pros_type=[]
pros_core=[]
rom=[]
ram=[]
pr_cam=[]
sec_cam=[]
battery=[]



for i in Url:
    driver.get(i)
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    time.sleep(1)
    model=driver.find_elements_by_xpath("//tr[@class='_1s_Smc row']")
    for j in model:
        text=j.find_element_by_xpath("td[1]").text
        if text=='Display Size':
            dis.append(j.find_element_by_xpath("td[2]").text)
        if text=='Resolution':
            res.append(j.find_element_by_xpath("td[2]").text)
            
            
        try:
            if text=='Processor Type':
                pros_type.append(j.find_element_by_xpath("td[2]").text)
            
        except:
            pros_type.append("-")
        
        if text=='Processor Core':
            pros_core.append(j.find_element_by_xpath("td[2]").text)
            
        if text=='Internal Storage':
            rom.append(j.find_element_by_xpath("td[2]").text)
            
        if text=='RAM':
            ram.append(j.find_element_by_xpath("td[2]").text)
            
        if text=='Battery Capacity':
            battery.append(j.find_element_by_xpath("td[2]").text)
            
            
        if text=='Primary Camera':
            pr_cam.append(j.find_element_by_xpath("td[2]").text)
            
        if text=='Secondary Camera':
            sec_cam.append(j.find_element_by_xpath("td[2]").text)
            
            
        
   
    
```

```python
pro=pd.DataFrame({})
pro['Brand Name']=Brand_Name[:19]
pro['Mobile Name']=Mobile_Name[:19]
pro['Model No']=Model_No[:19]
pro['Color']=Color[:19]
pro['Price']=Price[:19]
pro['Display size']=dis[:19]
pro['Resolution']=res[:19]
pro['Processor Type']=pros_type[:19]
pro['Processor core']=pros_core[:19]
pro['ROM']=rom[:19]
pro['RAM']=ram[:19]
pro['Primary camera']=pr_cam[:19]
pro['Secondary camera']=sec_cam[:19]
pro['Battery']=battery[:19]
pro
```
